$x(t)= A cos(\omega_0 t) + B sin(\omega_0 t)$

$\dot x(t)=\omega_0 [- Asin(\omega_0 t) + B cos(\omega_0 t)]$

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x=np.arange(0,10.1)
y=x**2
plt.plot(x,y)
```

```python
import matplotlib as mpl 
label_size= 14
mpl.rcParams["xtick.labelsize"]=label_size
mpl.rcParams["ytick.labelsize"]=label_size 
```

```python
a=100
w=.1
b=.1
t=np.linspace(0,50,1000)

plt.figure(figsize=(10,4))
plt.plot(t,a*np.cos(w*t)+ b *np.sin(w*t),'--*',lw=1 , ms = 5,label ='$x(t)$')
plt.plot(t,w*b*np.cos(w*t)-w*a*np.sin(w*t),'b-',lw = 1, ms= 7,label=r'$\dot {x(t)}$')
plt.xlabel ('$t$',fontsize=20)
plt.legend(loc = 1)
plt.show()
```
